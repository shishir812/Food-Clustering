```python
import pandas as pd
import numpy as np
import math
import operator
from collections import Counter
from matplotlib import pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score  
from sklearn.metrics import confusion_matrix 
dataset_new = pd.read_excel("Dataset.xlsx")
```

```python
dataset_new.rename(columns={'Favorite food names':'Favorite_food_names',
                            'Place of interest':'Place_of_interest'}, inplace=True)
columns = ['Place_of_interest','Favorite_food_names']
dataset_new_update = dataset_new[columns]
dataset_new_update
```

```python
dataset_new_update['Favorite_food_names'] = dataset_new_update['Favorite_food_names'].astype('category')
dataset_new_update['Favorite_food_names'] = pd.factorize(dataset_new_update['Favorite_food_names'])[0] + 1
dataset_new_update['Place_of_interest'] = pd.factorize(dataset_new_update['Place_of_interest'])[0] + 1
dataset_new_update
```

```python
D=dataset_new_update.values

# Get the labelled set
c1=D[:30,:]; c2=D[50:70,:];  c3=D[80:90,:]
trainSet = np.concatenate((c1,c2,c3),axis=0)

# Get the testing set
c1 = D[30:50,:]; c2=D[70:80,:];  c3=D[90:,:]
testSet = np.concatenate((c1,c2,c3),axis=0)

print(trainSet.shape)
print(testSet.shape)

xTrain=trainSet[:,:-1]; yTrain=trainSet[:,-1]
xTest=testSet[:,:-1]; yTest=testSet[:,-1]
```

```python
clf = KNeighborsClassifier(n_neighbors=5)
clf.fit(xTrain, yTrain.astype(int))
yPred=clf.predict(xTest)
acc=accuracy_score(yTest.astype(int), yPred.astype(int))
print('Accuracy with 5 neighbours: ',acc)
```

```python
def plot_conf_mat(lTrue, lPred, title):
    """ A function for plotting the confusion matrix given true and predicted labels."""
    cm = confusion_matrix(lTrue.astype(int), lPred.astype(int))
    print(cm)
    fig = plt.figure()
    ax = fig.add_subplot(111)
    cax = ax.matshow(cm)
    plt.title(title)
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.show()    
```

```python
plot_conf_mat(yTest, yPred, 'K=5')
```
